```python
class Abogado:
    def __init__(self, nombre, capacidad_analisis, comunicacion, etica):
        # Atributos (Datos del abogado)
        self.nombre = nombre
        self.capacidad_analisis = capacidad_analisis
        self.comunicacion_persuasiva = comunicacion
        self.etica_integridad = etica

    # acesoria acciones que realiza
    def asesoria(self):
        print(f"Abogado/a {self.nombre} está brindando asesoría legal: 'Basado en mi {self.capacidad_analisis}, le sugiero no firmar aún.'")

    def defensa(self):
        print(f"Abogado/a {self.nombre} está en plena defensa: 'Uso mi {self.comunicacion_persuasiva} para proteger los derechos de mi cliente.'")

    def gestion(self):
        print(f"Abogado/a {self.nombre} está gestionando documentos: 'Mi {self.etica_integridad} garantiza que este trámite es transparente.'")

# --- acesoria de objetos ---

# Obogado 1: Especialista en Civil
abogado1 = Abogado("Carlos García", "Alta (Estratega)", "Elocuente", "Inquebrantable")

# Abogado 2: Especialista en Penal
abogado2 = Abogado("Elena Torres", "Crítica y Rápida", "Directa", "Comprometida")

# --- acesoria de Funciones ---

print("--- Actividades del Abogado 1 ---")
abogado1.asesoria()
abogado1.defensa()

print("\n--- Actividades del Abogado 2 ---")
abogado2.gestion()
abogado2.defensa()
print("programa terminado")
print("clase abogado elaborado por Emily Barraza NL = 7 ")
```
